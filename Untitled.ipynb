```python
from flask import Flask, jsonify, request
import pandas as pd

app = Flask(__name__)

# Load the dataset
file_path = "C:/Users/kaushal/Downloads/final/100000 Sales Records.csv"  # Path to your CSV file
data = pd.read_csv(file_path)

# Convert DataFrame to JSON
data_json = data.to_dict(orient="records")

@app.route("/data", methods=["GET"])
def get_data():
    """Endpoint to fetch paginated data."""
    # Get query parameters
    page = int(request.args.get("page", 1))  # Default to page 1
    page_size = int(request.args.get("page_size", 100))  # Default to 100 rows per page
    
    # Calculate start and end indices
    start = (page - 1) * page_size
    end = start + page_size
    
    # Paginate the data
    paginated_data = data_json[start:end]
    
    # Return data as JSON
    return jsonify({
        "page": page,
        "page_size": page_size,
        "total_records": len(data_json),
        "data": paginated_data
    })

if __name__ == "__main__":
    app.run(debug=True, use_reloader=False)
```
